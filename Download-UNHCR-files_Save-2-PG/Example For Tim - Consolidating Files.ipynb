```python
import os 
import pandas as pd

input_folder = 'E:/UNHCR/temp/' 
#'E:\OneDrive - UNHCR/00 SharePoint Sites/Documents - Greening and Sustainability Team/Green Energy Team/3. Analysis/1. Load Imbalance Analysis/2. Source/TEMP/'

print(input_folder)
input_files = os.listdir(input_folder)



df_master = pd.DataFrame()

for i in input_files:
    print(i)
    df_temp = pd.read_excel(input_folder + i)
    df_master = pd.concat([df_master, df_temp])
    print(f"{i} has finished")

df_master
```

```python
# Load the consolidated dataframe to Postgres
from sqlalchemy import create_engine

engine = create_engine('postgresql://unhcr:unhcr@localhost:5432/unhcr')
print('Connection string created')

df_master.to_sql('file_consolidation_example', engine, if_exists = 'replace')
print('Data loaded to postgres')
```
